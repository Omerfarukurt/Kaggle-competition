```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv('train.csv')
```

```python
data.head(30)
```

```python
df=data.copy()
```

```python
df['datetime']=pd.to_datetime(df['datetime'])
```

```python
df.head()
```

```python
df['hour']=df['datetime'].dt.hour
```

```python
df.head()
```

```python
df['month']=df['datetime'].dt.month
```

```python
df.head()
```

```python
df.month.value_counts()
```

```python
df.drop(['registered','casual'],axis=1,inplace=True)
```

```python
df.drop('datetime',axis=1,inplace=True)
```

```python
df.head(25)
```

```python
plt.figure(figsize=(8,8))
sns.histplot(data=df, x='month', discrete=True)  # discrete=True aylara tam sayılar olarak davranır

plt.xlabel('Ay')
plt.ylabel('İşlem Sayısı (Count)')
plt.title('Aya Göre İşlem Sayısı')
plt.show()
```

```python
monthly_counts = df.groupby('month')['count'].sum().reset_index()

# Çubuk grafik çiz
sns.barplot(data=monthly_counts, x='month', y='count')
```

BURDA GÖRÜNENE GÖRE 5.AYDAN 10.AYA KADAR YANİ İLK VE SONBAHAR AYLARINDA KİRALAMA ORANI ARTMAYA BAŞLAMIŞ


```python
group_season=df.groupby('season')['count'].sum().reset_index()
```

```python
sns.barplot(data=group_season,x='season',y='count')
```

BURDAKİ VERİLERE GÖREDE KIŞIN KİRALANAN SAYI İLKBAHARI GEÇMİŞ      //1->İLKBAHAR   2->YAZ   3->SONBAHAR   4->KIŞ

```python
sns.histplot(data=df, x='month', hue='season', multiple='stack', discrete=True)
plt.show()
```
