```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv('train.csv')
```

```python
data.head(30)
```

```python
df=data.copy()
```

```python
df['datetime']=pd.to_datetime(df['datetime'])
```

```python
df.head()
```

```python
df['hour']=df['datetime'].dt.hour
```

```python
df.head()
```

```python
df['month']=df['datetime'].dt.month
```

```python
df.head()
```

```python
df.month.value_counts()
```

```python
df.drop(['registered','casual'],axis=1,inplace=True)
```

```python
df.drop('datetime',axis=1,inplace=True)
```

```python
df.head(25)
```

```python
plt.figure(figsize=(8,8))
sns.histplot(data=df, x='month', discrete=True)  # discrete=True aylara tam saylar olarak davranr

plt.xlabel('Ay')
plt.ylabel('lem Says (Count)')
plt.title('Aya Gre lem Says')
plt.show()
```

```python
monthly_counts = df.groupby('month')['count'].sum().reset_index()

# ubuk grafik iz
sns.barplot(data=monthly_counts, x='month', y='count')
```

BURDA GRNENE GRE 5.AYDAN 10.AYA KADAR YAN LK VE SONBAHAR AYLARINDA KRALAMA ORANI ARTMAYA BALAMI


```python
group_season=df.groupby('season')['count'].sum().reset_index()
```

```python
sns.barplot(data=group_season,x='season',y='count')
```

BURDAK VERLERE GREDE KIIN KRALANAN SAYI LKBAHARI GEM      //1->LKBAHAR   2->YAZ   3->SONBAHAR   4->KI

```python
sns.histplot(data=df, x='month', hue='season', multiple='stack', discrete=True)
plt.show()
```
